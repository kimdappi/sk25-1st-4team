### 데이터 불러오기

```python
import pandas as pd
genesis_df=pd.read_pickle("./data/FAQ_Genesis.pkl")
hyundai_df=pd.read_pickle("./data/FAQ_Hyundai.pkl")
kia_df=pd.read_pickle("./data/FAQ_KIA.pkl")
```

```python
# 세로로 합치기
faq_df = pd.concat(
    [genesis_df, hyundai_df, kia_df],
    axis=0,
    ignore_index=True
)
```
